```python
import os

# 대상 폴더 경로
folder_path = 'val/hsh_crop'

# 폴더 내 모든 파일 확인
for filename in os.listdir(folder_path):
    # 파일명이 원하는 패턴을 포함하는지 확인
    if 'frame_' in filename:
        # 'frame_' 이후 부분 제거
        new_name = filename.split('_frame_')[0] + ".png"  # 파일 확장자가 .png라고 가정
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, new_name)
        
        # 동일한 파일명이 존재할 경우 덮어쓰며 이름 변경
        os.replace(old_path, new_path)  # os.rename 대신 os.replace 사용
        print(f"Renamed (overwritten if exists): {filename} -> {new_name}")
```

```python
import os
import re

# 파일이 저장된 폴더 경로
folder_path = 'val/hsh_crop'

# 파일 이름 변경
for filename in os.listdir(folder_path):
    # 정규식을 사용하여 패턴 매칭
    match = re.match(r'(right_eye_img_\d+)_0*(\d+)', filename)
    if match:
        base_name = match.group(1)  # left_eye_img_1 부분
        number = int(match.group(2))  # 숫자를 정수로 변환 후 괄호 추가
        new_filename = f"{base_name}_({number})"
        
        # 파일 경로 생성
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, new_filename)
        
        # 파일 이름 변경
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} -> {new_filename}")
```
