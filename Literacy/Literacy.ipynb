## First Reading Data

```python
def get_all_csv_files(root_folder):
    csv_files = []
    for root, _, files in os.walk(root_folder):
        for file in files:
            if(file.endswith('.csv')):
                csv_files.append(os.path.join(root, file))
    return csv_files

def read_concat_csv(csv_files):
    dataframes = []
    for file in csv_files:
        df = pd.read_csv(file)
        folder_name = os.path.basename(os.path.dirname(file))
        df['folder_name'] = folder_name
        dataframes.append(df)
    concat_df = pd.concat(dataframes, ignore_index = True)
    return concat_df

root_folder = 'data/first_reading'
csv_files = get_all_csv_files(root_folder)

first_df = read_concat_csv(csv_files) # first_df -> first_reading_df : 처음 지문을 읽은 시선 좌표들

first_df
```
